```python
# adventure_game.py
# This script implements a text-based adventure game where a player navigates
# through different locations, makes decisions, and attempts to find a treasure.

print("Adventure Game Loaded Successfully!")
```

```python
def start_game():
    """
    Start the adventure game.
    - Introduce the treasure quest
    - Ask for player name
    - Offer a choice between 'forest' and 'cave'
    - Call forest_path() or cave_path() based on input
    """
    print("Welcome to the Adventure Game!")
    print("Your quest is to find the hidden treasure.\n")

    player_name = input("What is your name, adventurer? ").strip()
    print(f"\nHello, {player_name}! Your journey begins now.\n")

    while True:
        choice = input("Do you want to go to the 'forest' or the 'cave'? ").strip().lower()

        if choice == "forest":
            forest_path(player_name)
            break
        elif choice == "cave":
            cave_path(player_name)
            break
        else:
            print("Invalid choice. Please type 'forest' or 'cave'.\n")

def forest_path(player_name):
    """
    Forest scenario.
    - Describe entering a dark forest
    - Give the player two choices: follow a river or climb a tree
    - Use conditionals to determine the outcome
    """
    print(f"\n{player_name}, you enter a dark and mysterious forest.")
    print("You see a river flowing to your left and a tall tree to your right.\n")

    while True:
        choice = input("Do you want to 'follow the river' or 'climb the tree'? ").strip().lower()

        if choice == "follow the river":
            print("\nYou follow the river and discover a hidden treasure behind a waterfall!")
            end_game("win")
            return

        elif choice == "climb the tree":
            print("\nYou climb the tree but fall and injure yourself.")
            end_game("lose")
            return
        else:
            print("Invalid choice. Please type 'follow the river' or 'climb the tree'.\n")

def cave_path(player_name):
    """
    Cave scenario.
    - Describe entering a mysterious cave
    - Offer two choices: light a torch or proceed in the dark
    - Use conditionals to determine win or lose outcome
    """
    print(f"\n{player_name}, you enter a dark and eerie cave.")
    print("You can either light a torch to see or proceed in the dark.\n")

    while True:
        choice = input("Do you want to 'light a torch' or 'proceed in the dark'? ").strip().lower()

        if choice == "light a torch":
            print("\nYou light the torch and find a hidden passage leading to the treasure!")
            end_game("win")
            return

        elif choice == "proceed in the dark":
            print("\nYou stumble in the darkness and fall into a pit.")
            end_game("lose")
            return
        else:
            print("Invalid choice. Please type 'light a torch' or 'proceed in the dark'.\n")

def end_game(result):
    """
    End the game with a win/lose message and offer restart.
    """
    if result == "win":
        print("\n✅ You completed your quest and found the treasure!")
    elif result == "lose":
        print("\n❌ Your adventure ends here.")
    else:
        print("\nGame ended.")

    while True:
        again = input("\nWould you like to play again? (yes/no): ").strip().lower()

        if again in ("yes", "y"):
            print("\n--- Restarting the game ---\n")
            start_game()
            return
        elif again in ("no", "n"):
            print("\nThanks for playing!")
            return
        else:
            print("Invalid input. Please type 'yes' or 'no'.")

if __name__ == "__main__":
    start_game()
    

```
